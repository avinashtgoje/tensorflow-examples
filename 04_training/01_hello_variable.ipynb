# tf.nn.Variable

### a simple training operation. 

this is the deep learning equivalent of staring at the sun

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
```

#### x is a variable we are going to train

```python
some_example_random_number = 0.31415

x      = tf.Variable(some_example_random_number, dtype=tf.float32)
labels = tf.constant(1.0, dtype=tf.float32)
loss   = tf.reduce_sum(tf.square(x-labels))

train_op = tf.train.AdamOptimizer(learning_rate=0.5).minimize(loss)
init_op  = tf.global_variables_initializer()

with tf.Session() as sess:

    # initialize global variables
    sess.run(init_op)

    # run the training operation 
    for indx in range(20):
        sess.run(train_op)
        x_eval = sess.run(x)

        print("x = {}".format(x_eval))
```

let's visualize the training operation

```python
LEARNING_RATE = 0.5

some_example_random_number = 0.015

x      = tf.Variable(some_example_random_number, dtype=tf.float32)
labels = tf.constant(1.0, dtype=tf.float32)
loss   = tf.reduce_sum(tf.square(x-labels))

train_op = tf.train.AdamOptimizer(learning_rate=LEARNING_RATE).minimize(loss)
init_op  = tf.global_variables_initializer()

with tf.Session() as sess:

    # initialize global variables
    sess.run(init_op)

    # a place to store our trained value of x
    trained_x = []
    
    # run 100 steps the training operation 
    for indx in range(100):
        sess.run(train_op)
        trained_x.append( sess.run(x) )

    plt.plot(trained_x)
    plt.show()
```

wow, that was some crazy overshoot.

let's run this operation a few times with different learning rates

```python
learning_rates = [0.5, 0.4, 0.3, 0.2, 0.1, 0.05, 0.025]

for learning_rate in learning_rates:

    some_example_random_number = 0.015

    x      = tf.Variable(some_example_random_number, dtype=tf.float32)
    labels = tf.constant(1.0, dtype=tf.float32)
    loss   = tf.reduce_sum(tf.square(x-labels))

    train_op = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)
    init_op  = tf.global_variables_initializer()

    with tf.Session() as sess:

        # initialize global variables
        sess.run(init_op)

        # a place to store our trained value of x
        trained_x = []

        # run 100 steps the training operation 
        for indx in range(100):
            sess.run(train_op)
            trained_x.append( sess.run(x) )

        plt.plot(trained_x)
        plt.title("learning_rate={}".format(learning_rate))
        plt.show()
```
